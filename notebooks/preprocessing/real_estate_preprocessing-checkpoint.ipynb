```python
# add the furnished feature
df_temp['desc'].fillna("", inplace=True)
def extract_furnish(val):
    # if found furnished return true
    match = re.search(r"\bfurnished\b", val)
    if match:
        return True
    return False

# apply
df_temp['furnished_found'] = df_temp['desc'].apply(lambda x: extract_furnish(x))
df_temp['furnished_found'].value_counts()
```

```python
def mean_suburb_stats(df_list):
    """
    Args: 
        df_list: list of dataframes for each suburb within a postcode

    Returns:
        This function takes the sum of these dataframes vertically. This is 
        useful when a postcode is consisted of multiple suburbs and we want 
        to find the total number of bus stops for instance within this postcode:
         sum (bus stops in each suburb)
    """
    combined_df = pd.concat(df_list, axis=0)
    combined_df = combined_df[append_cols]

    # calculate the sum
    mean_df = pd.DataFrame(combined_df.sum(axis = 0, skipna = True)).transpose()
    # print(mean_df)
    return mean_df

def clean_suburb_name(name):
    """Removes all directions from a string.

    Args:
        string: The string to remove directions from.

    Returns:
        A string without directions.
    """
    directions = ['north', 'east', 'west', 'south']
    name_lst = [ele for ele in name.split(" ") if ele not in directions]
    
    return " ".join(name_lst)

def aggregate_liveability(df, live):
    """
    Args:
        df: rental real estate dataframe
        live: liveability df with features such as the number of bus stops

    Returns:

    This function merges the features in live dataframe into df based on the 
    postcode that this property locates

    This function is carefully designed for future extensibility as 1 postcode
    can map to many suburbs, leading to more complex feature handling

    Example of usage:
    (1)
    df: 599 elizabeth - 3053 - carlton
    live: carlton - 100 bus stops - 200 tram stops 
    Output: 599 elizabeth - 3053 - carlton - 100 bus stops - 200 tram stops 


    (2)
    df: 6 doll street  - 3996 - [inverloch, nhill]
    live: nhill - 100 bus stops - 200 tram stops 
          inverloch - 200 bus stops - 300 tram stops
    Output: 6 doll street-3996-[inverloch, nhill]-300 bus stops-500 tram stops 

    """
    # create previous values for new appended columns
    for col in append_cols:
        df[col] = None
    # iterate over each row
    for index, row in df.iterrows():
        
        curr_subs = ast.literal_eval(row['suburbs'])
        # remove any direction in the suburb
        curr_subs = [clean_suburb_name(sub) for sub in curr_subs]
        curr_subs = [string if 'st kilda' not in string 
                        else 'st kilda' for string in curr_subs]
        
        # 1 postcode maps to 1 suburb
        if len(curr_subs) == 1:
            appended_info = live[live['suburb'] == curr_subs[0]]
            # found
            for col in append_cols:
                df.loc[index, col] = appended_info[col].values[0]
        # more than one suburb found for each postcode
        else:
            # remove duplicates
            curr_subs = list(set(curr_subs))
            appended_info = [live[live['suburb'] == sub] for sub in curr_subs]

            # find the total number of stosp for each category within that
            # posecode
            info = mean_suburb_stats(appended_info)
            # append the new features
            for col in append_cols:
                df.loc[index, col] = info[col].values[0]
    return df
            
liveability = pd.read_csv(f"{CURATED_DATA_DIR}/liveability_features.csv")
append_cols = list(liveability.columns[1:])
df_temp = aggregate_liveability(df_temp, liveability)
df_temp.head()
```

## Distance to CBD
Find straight distance to CBD from each property

```python
#Loading Phoebe's PTV data
import geopandas as gpd


ptv_stops = gpd.read_file(f"{RAW_DATA_DIR}/ptv/metro_trains_stops.shp")
search_word = 'Melbourne'
result = ptv_stops[ptv_stops['stop_name'].str.contains(search_word, case=False)]

#Getting distance between 2 points in kilometers. 
# Let Melbourne Central PTV station coordinates represent CBD.
from geopy import distance
# location of mc: melbourne centarl
mc_coord = (ptv_stops['geometry'][17].y, ptv_stops['geometry'][17].x)

def distance_to_cbd(coords):
    # find straight distance calculated from coords to melbourne central 
    # coordinates
    return distance.distance(mc_coord, coords).km

# reformat the coords: lattitude and longitude for use of external libraries
for idx, row in df_temp.iterrows():
    lat_lon = row['coordinates'].strip('[]').split(', ')
    temp = []
    for num in lat_lon:
        temp.append(float(num))
    df_temp.at[idx, 'coordinates'] = temp
    
# calculate distance
df_temp['distance_to_cbd'] = df_temp['coordinates'].apply(distance_to_cbd)
mc_coord
```
