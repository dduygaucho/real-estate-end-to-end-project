```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
LANDING_DATA_DIR = "../../data/landing"
RAW_DATA_DIR = "../../data/raw"
CURATED_DATA_DIR = "../../data/curated"
```

```python
all_properties = pd.read_excel(f'{LANDING_DATA_DIR}/historical_rent.xlsx', sheet_name='All Properties').iloc[93][3: : 2]
flat_1br = pd.read_excel(f'{LANDING_DATA_DIR}/historical_rent.xlsx', sheet_name='1br flat').iloc[93][3: : 2]
flat_2br = pd.read_excel(f'{LANDING_DATA_DIR}/historical_rent.xlsx', sheet_name='2br Flat').iloc[93][3: : 2]
flat_3br = pd.read_excel(f'{LANDING_DATA_DIR}/historical_rent.xlsx', sheet_name='3br Flat').iloc[93][3: : 2]
house_2br = pd.read_excel(f'{LANDING_DATA_DIR}/historical_rent.xlsx', sheet_name='2br House').iloc[93][3: : 2]
house_3br = pd.read_excel(f'{LANDING_DATA_DIR}/historical_rent.xlsx', sheet_name='3br House').iloc[93][3: : 2]
house_4br = pd.read_excel(f'{LANDING_DATA_DIR}/historical_rent.xlsx', sheet_name='4br House').iloc[93][3: : 2]
```

```python
year_float = np.arange(1999.49, 2023.26, 0.25)
year = (np.floor(year_float)).astype(int)
month = ['June', 'September', 'December', 'March'] * 24
```

```python
median_rent = pd.DataFrame({'all_properties' : all_properties, \
                             'flat_1br' : flat_1br, \
                             'flat_2br' : flat_2br, \
                             'flat_3br' : flat_3br, \
                             'house_2br' : house_2br, \
                             'house_3br' : house_3br, \
                             'house_4br' : house_4br, \
                             'year' : year,\
                             'month' : month, \
                             'year_float': year_float})
median_rent = median_rent.reset_index(drop=True)
median_rent = median_rent[median_rent['year'] >= 2012]
median_rent = median_rent[median_rent['month'].isin(['March', 'September'])]
```

```python

for type in ['all_properties', 'flat_1br', 'flat_2br', 'flat_3br', 'house_2br', 'house_3br', 'house_4br']:
    plt.plot(median_rent['year_float'], median_rent[type], label = type.replace('_', ' '))

plt.xlabel("Year")
plt.ylabel("Median Rent")
plt.legend()
plt.title('Median Rent (Vic)')
plt.show()
```
